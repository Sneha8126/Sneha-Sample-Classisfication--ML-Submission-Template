## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x='Attrition',
    palette='Set2'
)

plt.title('Distribution of Employee Attrition', fontsize=15)
plt.xlabel('Employee Attrition', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

plt.show()
````

##### 1. Why did you pick the specific chart?

* A **count plot** is the most appropriate visualization for the target variable because Attrition is a categorical feature with two classes **(Stayed and Left)**.

* It clearly displays the frequency of employees in each category, making it easy to understand the distribution of the target variable and identify whether the dataset is **balanced or imbalanced**.


##### 2. What is/are the insight(s) found from the chart?

* The chart shows that the number of employees who Stayed is slightly higher than those who Left the company.

* The difference between the two classes is relatively small, indicating that the dataset is **reasonably balanced** and does not suffer from severe class imbalance.

* A significant number of employees have left the organization, highlighting that employee attrition is an important business concern that requires further investigation.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.


* Yes, the insights can create a positive business impact by helping the HR department understand the overall employee attrition level and prioritize retention strategies accordingly.

* Since a large number of employees have left the organization, the company can investigate the underlying causes and implement targeted improvements in areas such as employee satisfaction, career development, and workplace policies.

* A high attrition rate can negatively impact business growth because replacing experienced employees increases recruitment, onboarding, and training costs.

* Frequent employee turnover can also reduce productivity, disrupt team collaboration, and lead to the loss of valuable skills and organizational knowledge, ultimately affecting the company's overall performance and profitability.

#### Chart - 2

````python
# Chart - 2 visualization code
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    bins=20,
    kde=True,
    color='steelblue'
)

plt.title('Distribution of Employee Age', fontsize=15)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

plt.show()
````

##### 1. Why did you pick the specific chart?

* A histogram with a KDE (Kernel Density Estimation) curve was chosen because Age is a continuous numerical variable.

* This chart effectively displays the frequency distribution of employee ages while the KDE curve provides a smooth representation of the overall pattern.

##### 2. What is/are the insight(s) found from the chart?

* The employee age ranges from 18 to 59 years, indicating that the organization has a diverse workforce spanning both younger and experienced employees.

* The histogram shows that employees are distributed fairly evenly across most age groups, with no single age range dominating the workforce.

* There are slightly higher frequencies at the lower (around 18–20 years) and upper (around 57–59 years) age ranges, while the middle age groups remain relatively consistent.

* The KDE curve remains nearly flat throughout the age range, suggesting that the dataset has a well-balanced age distribution without significant skewness.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

* Yes, these insights can positively impact the business by helping HR understand the organization's workforce age composition and design age-specific engagement, training, and career development programs.

* The balanced age distribution indicates that the company has a healthy mix of young, mid-career, and experienced employees, which supports knowledge sharing and workforce stability.

* No direct negative growth is observed from this chart alone, as the age distribution appears balanced. However, if future analysis shows that a particular age group has a significantly higher attrition rate, it may indicate challenges related to career growth, employee engagement, or retention strategies for that group.

#### Chart - 3

````python
# Chart - 3 visualization code
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Attrition',
    y='Monthly Income',
    palette='viridis'
)

plt.title('Monthly Income by Employee Attrition', fontsize=16)
plt.xlabel('Employee Attrition', fontsize=12)
plt.ylabel('Monthly Income', fontsize=12)

plt.show()
````

##### 1. Why did you pick the specific chart?

* A box plot was selected because it is an effective visualization for comparing the distribution of a numerical variable (Monthly Income) across two categories of the target variable (Stayed and Left).


##### 2. What is/are the insight(s) found from the chart?

* The median monthly income of employees who Stayed and those who Left is very similar, indicating that salary levels are comparable between the two groups.

* Both groups contain several high-income outliers, showing that employees with higher salaries are present in both attrition categories.

* The significant overlap between the two box plots suggests that monthly income alone does not strongly differentiate employees who leave from those who stay.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

* Yes, these insights can create a positive business impact by helping management understand that increasing salaries alone may not be sufficient to reduce employee attrition.

* Since salary distributions are similar for both employees who stayed and those who left, HR should investigate additional factors that contribute to employee turnover instead of relying only on compensation.

* A negative business impact may occur if the organization assumes that salary is the only reason employees leave. Investing heavily in salary increases without addressing workplace satisfaction or growth opportunities may not significantly improve employee retention.

* These insights encourage a balanced retention strategy that combines competitive compensation with a positive work environment, ultimately improving employee satisfaction and organizational performance.

#### Chart - 4

````python
# Chart - 4 visualization code
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x='Job Role',
    hue='Attrition',
    palette=['#4C72B0', '#DD8452']
)

plt.title('Job Role vs Employee Attrition', fontsize=16)
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.xticks(rotation=45)

plt.show()
````

##### 1. Why did you pick the specific chart?

* A count plot is chosen because Job Role is a categorical variable, and the objective is to compare the number of employees **who Stayed and Left across different job roles.**

* This visualization makes it easy to identify which job roles have higher employee turnover and whether attrition varies significantly among different departments.

##### 2. What is/are the insight(s) found from the chart?

* The **Technology job** role has the highest number of employees, and it also records the highest number of employees who left the organization.

* **Healthcare and Education** also have a relatively high number of employees, with the number of employees who stayed being consistently higher than those who left.

* **Media and Finance** have comparatively fewer employees, and both roles also show lower attrition counts than the other job roles.

* Across all job roles, the number of employees who Stayed is higher than the number who Left, indicating that employee retention is generally stronger than attrition within each department.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

* Yes, these insights can create a positive business impact by helping HR identify job roles that require greater attention for employee retention and workforce planning.

* Since the Technology department has the highest number of employees leaving, management can further investigate factors such as workload, career growth opportunities, compensation, and job satisfaction within this department.

* Departments with lower employee turnover, such as Media and Finance, can be studied to identify effective workplace practices that may be implemented across other departments.

* **A high number of employees leaving from the Technology department may negatively impact business growth**, as it can lead to project delays, increased recruitment and training costs, and the loss of valuable technical expertise.
