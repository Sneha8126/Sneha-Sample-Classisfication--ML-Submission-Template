### 2. Handling Outliers

````python

# Step 1 :
num_columns = df.select_dtypes(include = ['int64','float64']).columns
print(num_columns)

# Step 2 :
fig, axes = plt.subplots(4, 2, figsize=(8, 6))
axes = axes.flatten()

for i, col in enumerate(num_columns):
    axes[i].boxplot(df[col], vert=False)
    axes[i].set_title(col, fontsize=8)
    axes[i].tick_params(axis='both', labelsize=6)

for j in range(len(num_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(pad=1)
plt.show()
````

````python
# Step 3 :
for col in num_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
````

````python
# Handling Outliers & Outlier treatments
df_outlier = df.copy()

# Columns having outliers
outlier_cols = ['Years at Company', 'Monthly Income']

for col in outlier_cols:
    Q1 = df_outlier[col].quantile(0.25)
    Q3 = df_outlier[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Cap the values
    df_outlier[col] = df_outlier[col].clip(lower=lower, upper=upper)

print("Outliers handled successfully.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were detected using the **Interquartile Range (IQR) method** and treated using **IQR-based capping (Winsorization)**. Instead of removing records, extreme values were capped to the IQR limits, which reduced the impact of outliers **while preserving all employee records**. This approach improves model performance and m**aintains the integrity of the dataset**.

````python
# ML Model - 1 Implementation

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    RocCurveDisplay
)

# Fit the Algorithm
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train, y_train)

# ===========================
# Train Accuracy
# ===========================
train_accuracy = lr.score(X_train, y_train)
print("Train Accuracy :", train_accuracy)
print()

# Predict on the model
y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)[:,1]

# ===========================
# Test Accuracy
# ===========================
test_accuracy = lr.score(X_test, y_test)
print("Test Accuracy :", test_accuracy)
print()

# Evaluation Metrics
print("Precision :", precision_score(y_test, y_pred_lr))
print()

print("Recall :", recall_score(y_test, y_pred_lr))
print()

print("F1 Score :", f1_score(y_test, y_pred_lr))
print()

cm = confusion_matrix(y_test, y_pred_lr)
print("Confusion Matrix")
print(cm)
print()

print("Classification Report")
print(classification_report(y_test, y_pred_lr))
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

from sklearn.model_selection import GridSearchCV

# Hyperparameter Grid
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs'],
    'max_iter': [500, 1000]
}

# GridSearchCV with 5-Fold Cross Validation
grid_search = GridSearchCV(
    estimator=LogisticRegression(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit the Algorithm
grid_search.fit(X_train, y_train)

# Best Model
best_lr = grid_search.best_estimator_

# Predict on the model
y_pred_best = best_lr.predict(X_test)

# Train Accuracy
train_accuracy = best_lr.score(X_train, y_train)

# Test Accuracy
test_accuracy = best_lr.score(X_test, y_test)

print("Best Parameters :", grid_search.best_params_)
print()

print("Best Cross Validation Accuracy :", grid_search.best_score_)
print()

print("Train Accuracy :", train_accuracy)
print()

print("Test Accuracy :", test_accuracy)
print()

print("Precision :", precision_score(y_test, y_pred_best))
print()

print("Recall :", recall_score(y_test, y_pred_best))
print()

print("F1 Score :", f1_score(y_test, y_pred_best))
print()

print("Confusion Matrix")
cm = confusion_matrix(y_test, y_pred_best)
print(cm)
print()

print("Classification Report")
print(classification_report(y_test, y_pred_best))
````

**GridSearchCV** was used as the **hyperparameter optimization technique** for the **Logistic Regression model**. It systematically evaluates all possible combinations of the selected hyperparameters using **5-fold Cross Validation** and identifies the combination that provides the **highest validation accuracy**.

The following hyperparameters were tuned:

* C (Regularization Strength)
* Solver
* Max Iterations (max_iter)
